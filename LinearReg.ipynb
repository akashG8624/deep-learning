```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/Salary_Data_1 (1).csv')
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X= df[['YearsExperience']]   # skill, experience  Independent variable
y= df['Salary'] # dependant 
```

```python
plt.figure(figsize=(4,3))
plt.xlim(0,12) 
plt.ylim(0,120000)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(X,y)
```

```python
model= LinearRegression()  # empty model
model.fit(X,y)
```

```python
model.intercept_
```

```python
yp= model.predict([[6]])   # prediction 
print(yp)
```

```python
yp=model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y,yp)
```

```python
mean_squared_error(y,yp)
```

```python
r2_score(y,yp)
```
